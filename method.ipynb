# Methods: function inside of a class is known as method.

# There are three types of methods:

1. class method
2. object method/instance method
3. static method

1. class method:- It is use to access and modify the class properties.

* classmethod decorator is use before  defining a class method.

* Symbol: #classmethod

* it take the cls first arguments

* cls carries the address of the class or object.

** Syntax: **
- class classname:
- @classmethod
- def func.name(cls)
- SB
- obj = classname()
- classname func_name()
- obj.func_name()


```python
class Bank:
    bname = "BOB"
    loc = "Noida"
    Ifsc = 'BOB123B'
    bmanager = 'Mr.Ashvik'

    @classmethod
    def banks_details(cls):
        print(cls.bname)
        print(cls.loc)
        print(cls.Ifsc)
        print(cls.bmanager)

    @classmethod
    def change_loc(cls):
        cls.loc = input("Enter the new location :")
        print("Location is change successfully :")
        cls.banks_details()




c1 = Bank()
c2 = Bank()
Bank.change_loc()

```

```python
class Student:
    # Class variable
    count = 0
    
    # Constructor (runs when object is created)
    def __init__(self, name, marks):
        self.name = name          # Instance variable
        self.marks = marks        # Instance variable
        Student.count += 1        # Increase count when a new object is created
    
    # Method to calculate grade
    def grade(self):
        if self.marks >= 90:
            return "A"
        elif self.marks >= 75:
            return "B"
        elif self.marks >= 60:
            return "C"
        else:
            return "D"

xhhhhhhhhhhhh
# Creating objects
s1 = Student("Ankit", 85)
s2 = Student("Rahul", 92)

# Printing details
print("Total Students:", Student.count)
print(s1.name, "Grade:", s1.grade())

```
