# Types of Properties/states/attributes/states/members/features--

There are two types of properties
1) class members- These are the members of the class which are same for all the objects
    - Example- schoolname,address,term&condition,principal

2) objects members- These are the property which are different for all the objects
    - Example- name,phonenumber,rollnumbers,adharcard,DOB.



    


```python
class demo:
    a=10
    b=20

obj1 = demo()
obj2 = demo()

#Accessing property  by using class 
print(demo.a)
print(demo.b)

# Accessing property by using objects

print(obj1.a)
print(obj1.b)
print('='*15)
print(obj2.a)
print(obj2.b)
print('='*15)
# modifying the properties 
demo.a = True
print(demo.a)
print(obj1.a)
print(demo.a)
print('='*15)
# Modiying the objects

obj1.b = False

print(obj1.b)
print(obj1.a)
print(obj1.b)

```

```python
class School:
    school_name = "KNIPSS"
    school_address = "Shrilanka"
    principal_name = "Jaychand"

# Creating objects
obj1 = School()
obj2 = School()
obj3 = School()

# Accessing properties using CLASS
print("Access Using Class:")
print(School.school_name)       
print(School.school_address)
print(School.principal_name)
print("=" * 15)

# Accessing properties using OBJECTS

print("Access Using Object:")
print(obj1.school_name)
print(obj1.school_address)
print(obj3.principal_name)
print("=" * 20)

# Modifying CLASS properties

School.school_name = "New KNIPSS"

print("After Modifying Class Attribute:")
print(School.school_name)
print(obj1.school_name)   
print(obj2.school_name)   
print("=" * 20)

#  Modifying OBJECT properties
obj1.school_address = "Delhi"         
obj1.principal_name = "Mr. Rohit"

print("After Modifying Object Attribute:")
print(obj1.school_address)       
print(obj1.principal_name)       
print(obj2.school_address)       
print(obj3.principal_name)       
```

# constructure / _ _ init _ _ /  ----------------

- It is a special method which is use to initilaize the members of the objects 
- We  don't need to call this method outside the class, it is invoked at the time of objects creation.
- It takes a mandatory arguments self , which carries the address of the objects.
- Syntax
    -  def _ _ _init_ _ _(self.arg1,arg2,arg3,----,argn)
        - self.property1 = arg1
        - self.property2 = arg3
        -   ---
       - self.property = arg3

```python
class Employee:
    def __init__(self, eid, ename, eadd, esal, ephono):
        self.eid = eid
        self.ename = ename
        self.eadd = eadd
        self.esal = esal
        self.ephono = ephono

        print(f"Employee with name {self.ename} is registered successfully...")


# Creating objects =======
e1 = Employee("E001", 'Ankit Verma', 'Sultanpur', 133323, '9999999999')
e2 = Employee("E002", 'Aman Sharma', 'Delhi', 15000, '8888888888')
e3 = Employee("E003", 'Rahul Singh', 'Lucknow', 18000, '7777777777')

#Store  in set =====
emp_list = {e1, e2, e3}

# Print in one line++++++++++++++++++++++++++
for emp in emp_list:
    print(emp.eid, emp.ename, emp.eadd, emp.esal, emp.ephono)
```

```python
class Showroom:
    def __init__(self, sid, sname, saddress, product_type, price, contact):
        self.sid = sid
        self.sname = sname
        self.saddress = saddress
        self.product_type = product_type
        self.price = price
        self.contact = contact


# Creating Showroom Objects----------------------------
s1 = Showroom("1001", "Ankit Motors", "Sultanpur", "Bikes", "200000", "87954561322")
s2 = Showroom("1002", "Aman Electronics", "Delhi", "Mobiles", "150000", "2315648975")
s3 = Showroom("1003", "Rahul Furniture", "Lucknow", "Furniture", "80000", "6548973215")


def show_details(showroom):

    print(showroom.sid,showroom.sname,showroom.saddress,showroom.product_type,showroom.price,showroom.contact)


# Displaying Showroom Details---------------------------------

show_details(s1)
show_details(s2)
#show_details(s2)
show_details(s3)


```
