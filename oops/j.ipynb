# ✅ oops concept-
   - OOPS = Object Oriented Programming System
   - Here everything is based on object that it is known as object orientend programming
   - To solve the real time problem oop concept is used.
   - (It is a programming style that uses objects and classes to structure code.)

- Why OOPS?
         - Code reusability
         - Easy to maintain
         - Makes real-world modeling possible
         - Reduces complexity



- ⭐ 1 Class-
          - class is a blueprint which contains the attributes and behaviour of an objects.
          - (A Class is a blueprint/template to create objects.)
          - class is a container where we can store the data and can access it. 

- ⭐ 2. Object-
          - It is a real time entity which have some attributes and properties.
          - (Object is the instance of a class.)    
          - It is an instance of class.
          - Object always belongs to a class.
          - We can create n number of object for a single class.
