# Multi level inheritance- 

- It is the phenamena of trasnferring property from single parent class to sigle child class by considering more than one level


```python


## ðŸ§  Syntax (Python)


class Grandparent:
    def gp_method(self):
        print("This is Grandparent class")

class Parent(Grandparent):
    def p_method(self):
        print("This is Parent class")

class Child(Parent):
    def c_method(self):
        print("This is Child class")

# Object creation
obj = Child()

# Calling methods
obj.gp_method()
obj.p_method()
obj.c_method()
```

# Multiple Inheritance
- In multiple inheritance the properties are transferred from from multiple parent class to single child class

   Parent1     Parent2
       \       /
        \     /
          Child


```python

## ðŸ§  Syntax (Python)

class Father:
    def father_method(self):
        print("This is Father class")

class Mother:
    def mother_method(self):
        print("This is Mother class")

class Child(Father, Mother):
    def child_method(self):
        print("This is Child class")

# Object creation
obj = Child()

# Calling methods
obj.father_method()
obj.mother_method()
obj.child_method()
```

# Hybrid Inheritance
- Hybrid Inheritance ek combination hota hai  
do ya do se zyada inheritance types ka.

ðŸ‘‰ Usually Multiple + Multilevel / Hierarchical ka mix hota hai.
- It is the combination of more than two inheritance

```python

# syntax
class A:
    def a_method(self):
        print("Class A method")

class B(A):
    def b_method(self):
        print("Class B method")

class C(A):
    def c_method(self):
        print("Class C method")

class D(B, C):
    def d_method(self):
        print("Class D method")

# Object creation
obj = D()

# Calling methods
obj.a_method()
obj.b_method()
obj.c_method()
obj.d_method()
```

```python
# Base class
class Math:
    a = "This is Math class"

# Single Inheritance
class Geometry(Math):
    b = "This is Geometry class"

# Multilevel Inheritance
class Calculator(Geometry):
    c = "This is Calculator class"

# Hierarchical Inheritance
class Chemistry(Calculator):
    d = "This is Chemistry class"

class Physics(Calculator):
    e = "This is Physics class"


# Object creation
chem = Chemistry()
phy = Physics()

# Accessing variables
print(chem.a)   # From Math
print(chem.b)   # From Geometry
print(chem.c)   # From Calculator
print(chem.d)   # From Chemistry

print("------")

print(phy.a)    # From Math
print(phy.b)    # From Geometry
print(phy.c)    # From Calculator
print(phy.e)    # From Physics
```

# 1) method overloading
- It is concept where same method contains different number of parameters 
- python does not support traditional overloading method like c++ and java but we can achive the method overloading
1) by using default argument
2) by using variable length 
    - variable length keyword args(*args,)

```python
class Arithmatic:
    @staticmethod
    def add(a,b):
        print(a+b)

    @staticmethod
    def add(a,b,c):
        print(a+b+c)

ob1 = Arithmatic()
ob1.add(10,20)
ob1.add(10,20,30)
```

```python
# By using default argument
class Arithmatic:
    @staticmethod
    def add(a,b,c=0):
        print(a+b+c)


ob1 = Arithmatic()
ob1.add(10,20)
ob1.add(10,20,30)
```

```python
# by using*args,**kwargs

class Arithmatic:
    @staticmethod
    def add(*args,**kwargs):
        print(sum(args))

ob1 = Arithmatic

ob1.add()
ob1.add(10,20)
ob1.add(10,20,30)
ob1.add(10,20,30,40, a=1,b=2,c=3)
```
