What is OOPS---

‚ú® OOPs stands for Object-Oriented Programming.
OOPs is a programming method where we model real-world things as objects that have properties and behaviors

üîπ Main Concepts of OOPs:
1) Class

1) Class-- Class is a blueprint which will store properties and functionlities.
in short,A class is like a blueprint of objects.

Create a class use the keyword class


2) Objects-  It is a real time entity 
or An object is an instance of a class.



```python
# Example of class--

# Define a class
class Student:
    # Constructor (used to initialize object data)
    def __init__(self, name, roll, marks):
        self.name = name
        self.roll = roll
        self.marks = marks

    # Method to display student details
    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Roll No: {self.roll}")
        print(f"Marks: {self.marks}")

# Create an object (instance) of the class
student1 = Student("Ankit", 101, 89)

# Call the method
student1.display_info()

```

```python
# EXample of objects--
class Student:
    def __init__(self, name, roll, marks):
        self.name = name
        self.roll = roll
        self.marks = marks

    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Roll No: {self.roll}")
        print(f"Marks: {self.marks}")
        print("-------------------------")

# Creating multiple objects
student1 = Student("Ankit", 101, 89)
student2 = Student("Suyash", 102, 92)
student3 = Student("Riya", 103, 85)

# Accessing data through methods
student1.display_info()
student2.display_info()
student3.display_info()
```
